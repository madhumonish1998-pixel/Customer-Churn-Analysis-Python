```python
# ================================
# Customer Churn Analysis Project
# ================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# Step 1: Create Sample Dataset
# -------------------------------

data = {
    "customer_id": range(1, 21),
    "tenure": [1, 24, 3, 48, 12, 36, 2, 60, 18, 5, 8, 22, 14, 7, 40, 10, 15, 4, 30, 6],
    "contract_type": [
        "Month-to-month","One year","Month-to-month","Two year","Month-to-month",
        "One year","Month-to-month","Two year","One year","Month-to-month",
        "Month-to-month","One year","Two year","Month-to-month","Two year",
        "One year","Month-to-month","Month-to-month","Two year","Month-to-month"
    ],
    "monthly_charges": [70,65,80,90,75,60,85,95,68,88,72,66,92,78,85,64,73,81,89,77],
    "total_charges": [70,1560,240,4320,900,2160,170,5700,1224,440,576,1452,1288,546,3400,640,1095,324,2670,462],
    "churn": [
        "Yes","No","Yes","No","Yes","No","Yes","No","No","Yes",
        "Yes","No","No","Yes","No","No","Yes","Yes","No","Yes"
    ]
}

df = pd.DataFrame(data)

# Display first 5 rows
print("First 5 Rows:")
print(df.head())

# -------------------------------
# Step 2: Churn Rate Calculation
# -------------------------------

churn_rate = df['churn'].value_counts(normalize=True) * 100
print("\nChurn Rate (%):")
print(churn_rate)

# -------------------------------
# Step 3: Churn Distribution Chart
# -------------------------------

plt.figure()
sns.countplot(x='churn', data=df)
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.show()

# -------------------------------
# Step 4: Contract Type vs Churn
# -------------------------------

plt.figure()
sns.countplot(x='contract_type', hue='churn', data=df)
plt.title("Churn by Contract Type")
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# Step 5: Correlation Heatmap
# -------------------------------

numeric_cols = df.select_dtypes(include=np.number)

plt.figure()
sns.heatmap(numeric_cols.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
